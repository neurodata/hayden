```python
np.random.seed(1)
reload_data = False

if reload_data:
    nodes = pd.read_csv('nodes', delimiter='\t')
    print(nodes.head())

    edges = pd.read_csv('edges', delimiter='\t',header=None)
    print(edges.head())

    G = nx.from_pandas_edgelist(edges, source=0, target=1)
    lcc = max(nx.connected_components(G), key=len)
    G_lcc = G.subgraph(lcc).copy()
    pkl.dump(G_lcc, open('nas_graph.pkl', 'wb'))
    
    N = len(G_lcc)
    in_sample_prop = 0.2

    nodes = pd.read_csv('nodes', delimiter='\t')['hash']

    in_sample_id = np.random.choice(nodes, int(N*0.2))

    temp_G = G_lcc.subgraph(in_sample_id).copy()
    lcc = max(nx.connected_components(temp_G))

    in_sample_G = temp_G.subgraph(lcc).copy()

    pkl.dump(in_sample_G, open('in_sample_G.pkl', 'wb'))
else:
#     G_lcc = pkl.load(open('nas_graph.pkl', 'rb'))
    in_sample_G = pkl.load(open('in_sample_G.pkl', 'rb'))
    
# print('largest connected component:' + str(len(G_lcc)))
print('in sample size:' + str(len(in_sample_G)))
```

```python
n, d = X_hat[:,:2].shape

fake_labels = -np.ones(n)
for i in range(n):
    if np.sqrt(np.sum(X_hat[i, :2]**2)) < 0.1:
        fake_labels[i] = 0
    elif X_hat[i, 0] > X_hat[i, 1]:
        fake_labels[i] = 1
    else:
        fake_labels[i] = 2
        
fig, ax = plt.subplots(1,1)
ax.set_xlabel('PC 1')
ax.set_ylabel('PC 2')
im = ax.scatter(X_hat[:, 0], X_hat[:, 1], c = fake_labels, s = 2)

plt.savefig('embedding_3class.png')
```

```python
X0 = X_hat[:, :2][fake_labels == 0].copy()
Y0 = rearranged_scores[fake_labels == 0].copy()

X1 = X_hat[:, :2][fake_labels == 1].copy()
Y1 = rearranged_scores[fake_labels == 1].copy()

X2 = X_hat[:, :2][fake_labels == 2].copy()
Y2 = rearranged_scores[fake_labels == 2].copy()
```

```python
def ecdf(x):
    xs = np.sort(x)
    ys = np.arange(1, len(xs)+1)/float(len(xs))
    return xs, ys

F0 = ecdf(Y0)
F1 = ecdf(Y1)
F2 = ecdf(Y2)

fig, ax = plt.subplots(1,1)
ax.plot(F0[0][1000:], F0[1][1000:], label="in B(0, 0.1)")
ax.plot(F1[0], F1[1], label="X1 > X2")
ax.plot(F2[0], F2[1], label="X2 > X1")
ax.set_xlabel('mean validation score')
ax.legend()
ax.set_title('ECDFs')
plt.savefig('nas_ecdfs_3class.png')
```

```python
good_classes = np.array(in_sample_ids)[fake_labels == 1]
bad_classes = np.array(in_sample_ids)[fake_labels == 2]
ball_classes = np.array(in_sample_ids)[fake_labels == 0]
pkl.dump(good_classes, open('good_classes.pkl', 'wb'))
```

```python
archs = pkl.load(open('in_sample_archs.pkl', 'rb'))

list_of_features = []
for i in range(archs.shape[0]):
    list_of_features.append([t for t in archs['node_features'].values[i][1:-1].replace("'", "").split(", ")])
```

```python
archs['lists of features'] = list_of_features
archs.head()
```

```python
layer_types = []
list_of_layer_counts = []
for list_ in list_of_features:
    temp_layer_count = np.zeros(len(layer_types))
    for ele in list_:
        if ele not in layer_types:
            layer_types.append(ele)
            temp_layer_count = np.concatenate((temp_layer_count, [1]))
        else:
            temp_idx = layer_types.index(ele)
            temp_layer_count[temp_idx] += 1
    list_of_layer_counts.append(temp_layer_count)
```

```python
count_matrix = np.zeros((archs.shape[0], len(layer_types) - 2))
for i in range(archs.shape[0]):
    temp = list_of_layer_counts[i]
    if len(temp) != len(layer_types):
        temp = np.concatenate((temp, np.zeros(len(layer_types) - len(temp))))
    count_matrix[i] = np.concatenate((temp[1:3], temp[4:]))
    
mid = [layer_types[1], layer_types[2], layer_types[4]]
layer_types
```

```python
archs[layer_types[1]] = count_matrix[:, 0]
archs[layer_types[2]] = count_matrix[:, 1]
archs[layer_types[4]] = count_matrix[:, 2]

archs['count vector'] = list(count_matrix[np.newaxis][0])

archs.head()
```

```python
good_archs = archs[archs.hash.isin(good_classes)].copy()
bad_archs = archs[archs.hash.isin(bad_classes)].copy()
ball_archs = archs[archs.hash.isin(ball_classes)].copy()

print('number of equivalence classes in "good" spire: {}'.format(len(good_classes)))
print('number of architectures in "good" spire: {}'.format(good_archs.shape[0]))
good_archs.head(11)
```

```python
good_counts = good_archs['count vector'].values
good_mat = np.zeros((good_counts.shape[0], 3))
for i, vec in enumerate(good_counts):
    good_mat[i] = vec
    
bad_counts = bad_archs['count vector'].values
bad_mat = np.zeros((bad_counts.shape[0], 3))
for i, vec in enumerate(bad_counts):
    bad_mat[i] = vec
    
ball_counts = ball_archs['count vector'].values
ball_mat = np.zeros((ball_counts.shape[0], 3))
for i, vec in enumerate(ball_counts):
    ball_mat[i] = vec
```

```python
good_lengths = np.sum(good_mat, axis = 1)
bad_lengths = np.sum(bad_mat, axis = 1)
ball_lengths = np.sum(ball_mat, axis = 1)

print('unique number of layers (and counts)')
np.unique(good_lengths, return_counts=True), np.unique(bad_lengths, return_counts=True), np.unique(ball_lengths, return_counts=True), 
```

```python
good_averages = np.mean(good_mat, axis = 0)
bad_averages = np.mean(bad_mat, axis = 0)
ball_averages = np.mean(ball_mat, axis = 0)

pkl.dump(good_averages, open('good_averages.pkl', 'wb'))
good_averages = pkl.load(open('good_averages.pkl', 'rb'))

pkl.dump(bad_averages, open('bad_averages.pkl', 'wb'))
bad_averages = pkl.load(open('bad_averages.pkl', 'rb'))

pkl.dump(ball_averages, open('ball_averages.pkl', 'wb'))
ball_averages = pkl.load(open('ball_averages.pkl', 'rb'))

good_averages, bad_averages, ball_averages
```

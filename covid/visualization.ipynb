```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt

A = np.random.standard_normal((10,10))

def build_graph(A):
    p, q = A.shape
    
    G = nx.Graph()
    G.add_nodes_from(np.arange(p*q))
    
    for i in range(p):
        for j in range(q):
            if i < p-1 and j < q-1:
                # edge right
                G.add_edge(i*p + j, i*p + j + 1)
                # edge down
                G.add_edge(i*p + j, (i+1)*p + j)
            elif i < p-1:
                # edge down 
                G.add_edge(i*p + j, (i+1)*p + j)
            elif j < q-1:
                # edge right
                G.add_edge(i*p + j, i*p + j + 1)
    return G
G = build_graph(A)

nx.draw_networkx(G, node_color=A.flatten(), cmap='bwr')
sm = plt.cm.ScalarMappable(cmap='bwr', norm=plt.Normalize(vmin = np.min(A), vmax=np.max(A)))
plt.colorbar(sm)
```
